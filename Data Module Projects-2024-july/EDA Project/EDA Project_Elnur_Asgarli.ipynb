```python
# 2. Perform basic Exploratory Data Analysis (EDA) on the data
# The structure of the data
print("Data Structure:")
print(df_kyp.info())

# Summary statistics
print("\nSummary Statistics:")
print(df_kyp.describe())

# Visualize the data
# Distribution of numerical columns
plt.figure(figsize=(10, 6))
sns.histplot(df_kyp['Age'], kde=True)
plt.title('Age Distribution')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df_kyp['Number'], kde=True)
plt.title('Number of Vertebrae Involved Distribution')
plt.show()

plt.figure(figsize=(10, 6))
sns.histplot(df_kyp['Start'], kde=True)
plt.title('Starting Vertebra Distribution')
plt.show()

# Count plot of the target variable
plt.figure(figsize=(10, 6))
sns.countplot(x='Kyphosis', data=df_kyp)
plt.title('Kyphosis Count')
plt.show()

# Check for missing values
print("\nMissing Values:")
print(df_kyp.isnull().sum())
```

```python
# 3. List the average, minimum and maximum age (in years) considered in this study using 2 methods
# Convert Age from month to years
age_years = df_kyp['Age']/12
# Method 1: using aggregate functions
print('Method 1')
print(f'Mean of Age: {age_years.mean():.2f}')
print(f'Minimum Age: {age_years.min():.2f}')
print(f'Maximum Age: {age_years.max():.2f}')
print('------------------------')
# Method 2: using descriptive statistics
age_stat = age_years.describe()
print('Method 2')
print(f'Mean of Age: {age_stat["mean"]}')
print(f'Minimum Age: {age_stat["min"]}')
print(f'Maximum Age: {age_stat["max"]}')
```

```python
# 4. Plot the correlation matrix
# 'Kyphosis' column is categorical with absent and present values that will give error if corr() function is directly applied
# Need to convert it to labels before finding correlation. For this, we will import LabelEncoder
from sklearn.preprocessing import LabelEncoder

# Convert Kyphosis column to numerical
le = LabelEncoder()

df_kyp['Kyphosis'] = le.fit_transform(df_kyp['Kyphosis'])

# Calculate the correlation matrix
corr_matrix = df_kyp.corr()
corr_matrix
```

```python
# Correlation matrix heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_kyp.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

Explanation of the relationsips:
- Kyphosis vs. Age: 0.126465.
There is a weak positive correlation between Kyphosis and Age. This means that there is a very slight tendency for older ages to be associated with kyphosis presence, but the correlation is quite weak.

- Kyphosis vs. Number: 0.360935.
There is a moderate positive correlation between Kyphosis and the Number of vertebrae involved in the surgery. This suggests that surgeries involving more vertebrae are somewhat more likely to result in kyphosis.
- Kyphosis vs. Start: -0.445943.
There is a moderate negative correlation between Kyphosis and the Start position of the vertebra operated on. This means that starting the surgery on higher vertebrae (closer to the head) is associated with a lower likelihood of kyphosis.

```python
#  5. Convert the age column datatype from int64 to float64
df_kyp['Age']=df_kyp['Age'].astype('float64')
df_kyp.info()
```

```python
# 6. Define a function that converts age from months to years
def age_convert(age_in_month):
    age_in_years = age_in_month/12
    return age_in_years
```

```python
# 7. Apply the function to the Age column and add the results into a new column entitled Age in Years 
df_kyp['Age in Years'] = df_kyp['Age'].apply(age_convert)
df_kyp.head()
```

```python
# 8. What are the features of the oldest and youngest child in this study?
# Find the youngest and oldest child
youngest_child = df_kyp.loc[df_kyp['Age'].idxmin()]
oldest_child = df_kyp.loc[df_kyp['Age'].idxmax()]

print("Features of the youngest child:\n", youngest_child)
print("\nFeatures of the oldest child:\n", oldest_child)
```

```python
# 9. Scale the Raw Age Column Using Standardization and Normalization
# Standardization and normalization are two different techniques to scale data. Let's perform both on the Age column.

# Standardization
# Standardization scales the data to have a mean of 0 and a standard deviation of 1.

# Normalization
# Normalization scales the data to be within a specific range, typically 0 to 1.
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Standardization
scaler_standard = StandardScaler()
df_kyp['Age_standardized'] = scaler_standard.fit_transform(df_kyp[['Age']])

# Normalization
scaler_minmax = MinMaxScaler()
df_kyp['Age_normalized'] = scaler_minmax.fit_transform(df_kyp[['Age']])

# Sanity check
print("\nStandardized Age:\n", df_kyp[['Age', 'Age_standardized']].head())
print("\nNormalized Age:\n", df_kyp[['Age', 'Age_normalized']].head())
```

Separation of Diagnosis:

By coloring the points based on the diagnosis (Malignant in red and Benign in blue), we can observe how the two categories distribute across the mean radius and mean area.
Malignant tumors often have larger and more irregular nuclei, so we see a tendency for red points to cluster at higher values of mean radius and mean area.

Cluster Formation:

The scatter plot shows clusters of points that indicate typical ranges for benign and malignant tumors.
Benign tumors (blue points) are clustered around lower values of mean radius and mean area, while malignant tumors (red points) are clustered around higher values.

```python
# Import the required librarires
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

```python
# Import the datasets
stock_prices = pd.read_csv('stocks_daily_prices.csv')
stock_returns = pd.read_csv('stocks_daily_returns.csv')
```

```python
# Plot lineplots of all 4 stocks daily prices on one single figure
fig, ax = plt.subplots(figsize=(16, 6))  # Create a new figure with the desired size
stock_prices[['JPM', 'PG', 'AAPL', 'UAL']].plot(ax=ax, linewidth=1)
plt.title('Daily Stock Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Plot 4 stocks daily prices on multiple subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
stock_prices[['JPM']].plot(ax=axes[0, 0], linewidth=2)
axes[0, 0].set_title('JP Morgan Chase')
stock_prices[['PG']].plot(ax=axes[0, 1], linewidth=2)
axes[0, 1].set_title('Procter & Gamble')
stock_prices[['AAPL']].plot(ax=axes[1, 0], linewidth=2)
axes[1, 0].set_title('Apple')
stock_prices[['UAL']].plot(ax=axes[1, 1], linewidth=2)
axes[1, 1].set_title('United Airlines')
plt.tight_layout()
plt.show()
```

```python
# Plot the 4 plots on subplots next to each other (all figures in one row)
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
stock_prices[['JPM']].plot(ax=axes[0], linewidth=2)
axes[0].set_title('JP Morgan Chase')
stock_prices[['PG']].plot(ax=axes[1], linewidth=2)
axes[1].set_title('Procter & Gamble')
stock_prices[['AAPL']].plot(ax=axes[2], linewidth=2)
axes[2].set_title('Apple')
stock_prices[['UAL']].plot(ax=axes[3], linewidth=2)
axes[3].set_title('United Airlines')
plt.tight_layout()
plt.show()
```

```python
# Plot the scatterplot between Apple and JP Morgan daily returns
plt.figure(figsize=(8, 6))
plt.scatter(stock_returns['AAPL'], stock_returns['JPM'])
plt.xlabel('Apple Daily Returns')
plt.ylabel('JP Morgan Daily Returns')
plt.title('Scatterplot of Apple and JP Morgan Daily Returns')
plt.show()
```

```python
# Plot the scatterplot between Apple and JP Morgan daily returns using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x=stock_returns['AAPL'], y=stock_returns['JPM'])
plt.xlabel('Apple Daily Returns')
plt.ylabel('JP Morgan Daily Returns')
plt.title('Scatterplot of Apple and JP Morgan Daily Returns (Seaborn)')
plt.show()
```

```python
# Assume that you decided to become bullish on AAPL and you allocated 70% of your assets in it. 
# You also decided to equally divide the rest of your assets in other stocks (JPM, PG, and UAL). 
# Using Matplotlib, plot a pie chart that shows these allocations. Use 'explode attribute to 
# increase the separation between AAPL and the rest of the portfolio.
# Plot a pie chart with the given asset allocation
allocation = [0.7, 0.1, 0.1, 0.1]
labels = ['AAPL', 'JPM', 'PG', 'UAL']
explode = (0.1, 0, 0, 0)
plt.figure(figsize=(8, 6))
plt.pie(allocation, labels=labels, explode=explode, autopct='%1.1f%%')
plt.title('Portfolio Allocation')
plt.axis('equal')
plt.show()
```

```python
# Using Matplotlib, plot the histogram for United Airlines and P&G returns using 40 bins with red color. 
# Display the mean and Standard deviation for both stocks on top of the figure. 

plt.figure(figsize=(10, 6))
stock_returns['UAL'].plot(kind='hist', bins=40, color='red', alpha=0.5)
stock_returns['PG'].plot(kind='hist', bins=40, color='blue', alpha=0.5)

# Add vertical lines for mean and standard deviation
plt.axvline(stock_returns['UAL'].mean(), color='r', linestyle='dashed', linewidth=1)
plt.axvline(stock_returns['UAL'].std(), color='r', linestyle='dashed', linewidth=1)
plt.axvline(stock_returns['PG'].mean(), color='k', linestyle='dashed', linewidth=1)
plt.axvline(stock_returns['PG'].std(), color='k', linestyle='dashed', linewidth=1)

plt.text(-40, 1250, f"UAL Mean: {stock_returns['UAL'].mean():.4f}", color='red')
plt.text(-40, 1200, f"UAL Std: {stock_returns['UAL'].std():.4f}", color='red')
plt.text(-40, 1100, f"PG Mean: {stock_returns['PG'].mean():.4f}", color='blue')
plt.text(-40, 1050, f"PG Std: {stock_returns['PG'].std():.4f}", color='blue')

plt.title('Histogram of United Airlines and P&G Returns')
plt.xlabel('Returns')
plt.ylabel('Frequency')
plt.legend(['UAL', 'PG'])
plt.show()
```

```python
# What did you conclude from this graph?
# Both graphs seem to be normally distributed while UAL returns have a wider standard deviation 
# that means stock returns for United Airlines have been more volatile.
```

```python
# Plot a heatmap showing the correlations between stocks daily returns
plt.figure(figsize=(8, 6))
sns.heatmap(stock_returns[['AAPL','JPM','PG','UAL']].corr(), annot=True)
plt.title('Correlation Heatmap of Stocks Daily Returns')
plt.show()
```

```python
# Plot a 3D plot showing all daily returns from JPM, AAPL and UAL
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(stock_returns['JPM'], stock_returns['AAPL'], stock_returns['UAL'])
ax.set_xlabel('JP Morgan Daily Returns')
ax.set_ylabel('Apple Daily Returns')
ax.set_zlabel('United Airlines Daily Returns')
plt.title('3D Plot of Daily Returns')
plt.show()
```
