## Unique Values

```python
# To get detailed information about the data set;

def get_unique_values(df):
    
    output_data = []

    for col in df.columns:

        # If the number of unique values in the column is less than or equal to 10
        if df.loc[:, col].nunique() <= 10:
            # Get the unique values in the column
            unique_values = df.loc[:, col].unique()
            # Append the column name, number of unique values, unique values, and data type to the output data
            output_data.append([col, df.loc[:, col].nunique(), unique_values, df.loc[:, col].dtype])
        else:
            # Otherwise, append only the column name, number of unique values, and data type to the output data
            output_data.append([col, df.loc[:, col].nunique(),"-", df.loc[:, col].dtype])

    output_df = pd.DataFrame(output_data, columns=['Column Name', 'Number of Unique Values', ' Unique Values ', 'Data Type'])

    return output_df
```

# ML

**ML Steps:**

**1.** First, we assign the target as y and the features as X. 

**2.** Secondly, the data set is divided (split) into Train and Test. (train_test_split; min 70% will be train, 30% test)

**3.** Third, the Model is defined. Since we will use the "linear regression model" here, we will do this as "model=LinearRegression()".

**4.** The data is scaled like "scaler = StandardScaler()". The scaler is trained on train data using "scaler.fit(X_train)". This scaling process is transformed into train and test data like "X_train_scaled = scaler.transform(X_train)"; "X_test_scaled = scaler.transform(X_test)".

**5.** The model is fitted using "X-train_scaled" and "y_train" like "model.fit(X_train_scaled, y_train)".

**6.** "y_train_pred" and "y-predict" are obtained using the "y_train_pred = model.predict(X_train_scaled)" and "y_pred = model.predict(X_test_scaled)" respectively. 

While doing this, we obtain y_predict by giving the X_test data to the regression model (y=aX1+bX2+cX3+dX4+e). We find a,b,c,d,e here by calling coefficients and intercepts.

**7.** Then we compare "y_predict" with "y_actual", that is "y_test". We obtain the residues. We measure the success level of our model by applying eval metrics tests (eval_metric) on residuals.

**8.** We also obtain the error metrics of the "train data" and compare them with the evaluation metrics we obtained from the "test data."

**9.** We try different models and obtain their evaluation metrics. We then look at the evaluation metrics to determine the most successful model. We name this model the "final_model." We train this model using the entire data set and "save" that model.

## Scaling the Data

* When the sizes of the data we will use in the analysis differ size, the **model tends to give more importance to the larger numbers**. To **eliminate this problem, scaling** is applied to make the data the same size.

* We will **fit the scaling on train** data and **transform it to train and test** data.

* By doing this, the **data leakage** problem is avoided. 

```python
y_train.head(3)
```

```python
X_train.head(3)
```

* There are big difference between CO2 emission and engin_size. 

```python
#Let's import the libraries we will use in the scaling process;

from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
scaler = StandardScaler() # There are 3 options here; we choose this since we will standardize.
                          # MinMaxScaler is used for normalization.
```

```python
# The scaling function learns how to scale (standardize) through "X_train" data by "fit". 

scaler.fit(X_train)  
```

```python
# Applies the scaling system learned above to the X_train data with "transform".

X_train_scaled = scaler.transform(X_train)  
X_train_scaled 
```

!!! **NOT:** While we apply both **fit** and transform to the "**X_train**" data, only **transform** is applied to the "**X_test**" data. 

!!! **Scaling** operations cannot be applied to **y_train** and **y_test** data.!!!!!!!!** 

```python
# In order to scale the "X_test" series, we will "only transform" it.
# Thus, while scaling X_test, it uses the information it learned with "fit" on X_train.

X_test_scaled = scaler.transform(X_test) 
X_test_scaled 
```

* When the X_train and X_test series, which contain large numbers in their raw form, were "standardized," their values decreased. Their mean became "0," and their standard deviation became 1.

* Now we have scaled **X_train_scaled** and **X_test_scaled** data. 

## SIMPLE LIENAR REGRESSION  

y = b0 + b1 * X 

CO2 = b0 + b1 * "engine_size"  

```python
# Let's import the necessary library to estimate the Linear Regression model;

from sklearn.linear_model import LinearRegression
```

```python
# Let's assign the library to an object (we named it "lin_reg", we can give it any name) (for practicality and so we can remember it);

lin_reg = LinearRegression()
```

```python
# Let's estimate (fit) the model (We obtain the coefficients (b0, b1) of the equation); 

lin_reg.fit(X_train_scaled, y_train)
```

```python
# Let's see the intercept (b0) of the model.
lin_reg.intercept_ # b0
```

```python
# Let's see the slope (b1) of the model.
lin_reg.coef_      # b1
```

```python
# Let's assign the intercept and slope coefficient to b0 and b1 (so I can use them when performing manual operations below).
b0 = lin_reg.intercept_
b1 = lin_reg.coef_
```

* If we write these coefficients on the linear regression model, we obtain the following model:

y = 250.93 + 49.76 * X ...............................................................(1)

CO2 = 250.93  + 49.76 * "engine_size" .......................................(2)

**How can we interpret results of the Simple Linear Regression Model?**

**For b0:**
* Regardless of "Engine Size", cars have been found to emit an average of "250.93 g/km of CO2".

**For b1:**
* If X1 increase 1 unit, y will increse b1 unit.
**If "engine_size" increase 1 liter, CO2 emisson of that car will incerease 49.76 g/km**.

### Let's Apply the Obtained Model to the "X_train" Data

* We can calculate the values of CO2 emission (**y_pred_train**) by writing X_train_scaled  data on Eq (2). 

```python
# Manuel;

y_train_pred =  b0 + b1 * X_train_scaled
y_train_pred
```

```python
# By code;

y_train_pred = lin_reg.predict(X_train_scaled)
y_train_pred
```

* We found this value to **test the success level of the model**.

* **We examine the success of the model** by using **eval_metrics**".

### eval_metrics for "Train Data" 

```python
# Let's call the necessary library for eval metrics;

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# Train Datasi icin eval metricsleri elde edelim; 

def eval_metric(actual, pred):
    mae = mean_absolute_error(actual, pred)
    mse = mean_squared_error(actual, pred)
    rmse = np.sqrt(mean_squared_error(actual, pred))
    R2_score = r2_score(actual, pred)
    #print("Model testing performance:")
    print("---------------------------------------")
    print(f"R2_score \t: {R2_score}")
    print(f"MAE \t\t: {mae}")
    print(f"MSE \t\t: {mse}")
    print(f"RMSE \t\t: {rmse}")
    
print("Model testing performance for TRAIN DATA:")  




eval_metric(y_train, y_train_pred)
```

### Let's Apply the Estimated Model to the "Test" Data

* Let's test the model we obtained using "Test" data.

* For this, we write the X_test_scaled data instead of X in the model;

```python
# Manuel;

y_pred = b1 * X_test_scaled + b0
y_pred
```

```python
# By code;

y_pred = lin_reg.predict(X_test_scaled)
y_pred
```

* These values are the estimated CO2 emission (i.e. **y^**) values of the test section.

### eval_metrics for "Test Data"

```python
# Let's obtain eval metrics for Test Data; 

eval_metric(y_test, y_pred)
```

### Let's comparatively evaluate the prediction results we made for Train and Test data using "eval metrics" 

```python
print("Simple Linear Model testing performance for TRAIN DATA:")
print(eval_metric(y_train, y_train_pred))

print("---------------------------------------")

print("Simple LinearModel testing performance for TEST DATA:")
print(eval_metric(y_test, y_pred))
```

#### 2nd (Short) Way for Eval Metrics

We can also examine Eval metrics side by side with the codes below.

```python
def train_val(y_train, y_train_pred, y_test, y_pred, i):
    
    scores = {i+"_train": {"R2" : r2_score(y_train, y_train_pred),
    "mae" : mean_absolute_error(y_train, y_train_pred),
    "mse" : mean_squared_error(y_train, y_train_pred),                          
    "rmse" : np.sqrt(mean_squared_error(y_train, y_train_pred))},
    
    i+"_test": {"R2" : r2_score(y_test, y_pred),
    "mae" : mean_absolute_error(y_test, y_pred),
    "mse" : mean_squared_error(y_test, y_pred),
    "rmse" : np.sqrt(mean_squared_error(y_test, y_pred))}}
    
    return pd.DataFrame(scores)
```

```python
simple_linear = train_val(y_train, y_train_pred, y_test, y_pred, "simple")
simple_linear
```

**DEGERLENDIRME:** **1.**  According to these results, the **model's success level is 72%** in both train data and test data. In other words, **we can explain 72% of the change in CO2 emissions by using "engine_size"** data. In other words, we have enough data to explain 72% of the changes in CO2.

**2.** The model's explanatory power (R2 = 0.72) is not very high.

**3.** Since there is no significant difference between the R2 values of the train and test analyses and other test statistics (R2 in the analysis with test data is not much smaller than the R2 obtained in the analysis with train data), there is NO OVERVITTING problem in the model.

**4.** The fact that the MAE and RMSE values of the models are relatively far from each other indicates that there are **outliers** in the data set. 

## Cross-Validation

Let's perform cross-validation to see whether the eval metrics (R2, MAE, MSE, RMSE) we obtained in the train section are the same as all train data.



```python
# Let's import the necessary libraries to perform Cross-Validation;

from sklearn.model_selection import cross_validate, cross_val_score
```

```python
# Let's assign the Linear Regression model to the "model" object and obtain the scores of this model;
model = LinearRegression() # Here we can also use "lin_reg = LinearRegression()". 
scores = cross_validate(model, X_train_scaled, y_train, scoring = ['r2', 'neg_mean_absolute_error','neg_mean_squared_error', \
                                                            'neg_root_mean_squared_error'], cv = 10) # cv; cross validation 

# NOTE !!!
# Note that the error metrics are prefixed with "negative". This is to make the scores closer to 0. When interpreting, 
# we will interpret them without a sign (as if they were positive).
```

```python
# Let's see the scores obtained in each step of the Cross-Validation process together;

pd.DataFrame(scores, index = range(1,11)) # 0 dan başlatmasın diye 1 verdim indexe. Sadece gorsellik acisindan. 
```

```python
# Let's calculate the average scores;

scores = pd.DataFrame(scores, index=range(1,11)) 

# Update column names 
scores.columns = [col.replace('test_', '') for col in scores.columns]

# Average scores
mean_scores = scores.iloc[:, 2:].mean()
mean_scores
```

**Conclussions:** The R2 we obtained after CV was 0.72. This is **consistent with the R2 we obtained with the train data**. The **MAE and RMSE** values, which we will interpret as unsigned as a result of the CV process, are also **consistent with the results we obtained for all train data**. Then the eval metrics obtained are valid for all train data and there isn't an OVERFITTING or UNDERFITTING problem in our model since scores are consistent.

It's essentially a diagonal line at a **45-degree angle**, plotted along with the scatter points of your actual (y) vs. predicted (y^) values.

**Interpretation:**

**Points close to the identity line:** These indicate **good model performance**, as the **predictions are very close to the true values**.

**Points deviating from the identity line:** The further **away a point** is, the **larger the prediction error** for that particular instance. This could reveal areas where the **model needs improvement**.

**Benefits of Analyzing the Identity Line:**

**Visual Assessment of Prediction Accuracy:** The identity line provides a quick visual reference for gauging how well your model performs on unseen test data.

**Identifying Biases or Patterns:** Deviations from the line can indicate potential biases or non-linear relationships that the model might not be capturing effectively.

**Additional Considerations:**

The identity line **serves as a theoretical best-case scenario**, but achieving **perfect prediction is usually unrealistic** for real-world regression problems.

The presence of noise or inherent variability in the data will naturally lead to some scatter around the line.

## Multicollinearity Check 

Was there a **Multicollinearity Problem** between the data? We can examine this by examining the **correlation** between the **features** (explanatory variables). When the **correlation is > 0.90**, the "**Multicollinearity**" problem **may occur**.

**NOTE:** Even if this problem occurs, there is **no** need to **panic**. You will learn **REGULASITION methods** (such as L1 and L2), which **will solve this problem as well**. 

```python
def color_correlation2(val):
    """
    Takes a scalar and returns a string with
    the css property in a variety of color scales 
    for different correlations.
    """
    if val >= 0.9 and val < 0.99999 or val <= -0.6 and val > -0.99999:
        color = 'red'
    elif val < 0.9 and val >= 0.3 or val > -0.6 and val <= -0.3:
        color = 'blue'
    elif val == 1:
        color = 'green'    
    else:
        color = 'black'
    return 'color: %s' % color
 
df.corr().style.applymap(color_correlation2)
```

## Extra Metrics for Calculating the Accuracy Value of the Model

**1)** **RATIO = RMSE/AVG_TARGET**

* RATIO < 0.10: The model is considered **successful**. The error rate is less than 10% of the average value of the target variable, indicating **very good model performance**.

* RATIO 0.10 - 0.20: The model performance is **acceptable**. The error rate is between 10% and 20% of the average value of the target variable.

* RATIO > 0.20: The model **performance is poor**. The error rate is greater than 20% of the average value of the target variable, indicating **significant prediction errors**.


It may be **wrong** to comment on the model's success using only this ratio in data sets with many **outliers**. **This ratio should be evaluated with other metrics (R², MAE)**.


**2)** We can move one step more by using this formula: 
  **Accuracy_Ratio = ((1 - RATIO) * 100).round(1)**
  We can interpret it as a succes level based on 100.
  
 * The **RATIO** represents the **relative error**. By subtracting the RATIO from 1 and multiplying by 100, you **convert this error** into an **accuracy percentage**.

* A **lower RATIO** and **higher Accuracy_Ratio** indicate a **higher accuracy rate**.

The **Accuracy Rate** provides a **more intuitive understanding** of the **model's performance** by showing the **percentage of accuracy**. 

This can be **particularly useful** for **communicating model performance** to **stakeholders** who **may find percentages easier to interpret** than raw error metrics.

**!! We should present our results more easily to non-technical audiences.**
