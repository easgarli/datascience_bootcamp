## Import Libraries

```python
import pandas as pd 

# To ignore/hide the warnings
import warnings
warnings.filterwarnings("ignore")
```

## Read Data

```python
df = pd.read_csv("wineData.csv", encoding="latin-1")
df.head(3)
```

```python
df_Paul_Jaboulet_Aîné = df[df['winemaker'] == 'Paul Jaboulet Aîné']
df_Paul_Jaboulet_Aîné
```

```python
df_Paul_Jaboulet_Aîné.shape
```

```python
df_Henry_Fessy = df[df['winemaker'] == 'Henry Fessy']
df_Henry_Fessy
```

```python
df_Henry_Fessy.shape
```

```python
df_Paul_Jaboulet_Aîné['wine_score']
```

```python
df_Henry_Fessy['wine_score']
```

**Hypotheses:**




```python
# To test these hypotheses, we use the "Two sample independent t-test."
from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(df_Paul_Jaboulet_Aîné['wine_score'], df_Henry_Fessy['wine_score'], alternative='greater')

print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")
```

```python
# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: df_Paul_Jaboulet_Aîné['wine_score'] mean is greater than df_Henry_Fessy['wine_score'] mean.")
else:
    print("Fail to reject the null hypothesis: No evidence that df_Paul_Jaboulet_Aîné['wine_score'] mean is greater than df_Henry_Fessy['wine_score'] mean.")
```

**RESULT :** Since p_value = 0.0000097 < 0.05, **the Null hypothesis was rejected, and a decision was made that the wines of Paul Jaboulet Aîné have higher quality than the wines of Henry Fessy**.
