```python
# Practical Way
from scipy import stats

# Data for the two catalysts
catalyst_1 = [91.50, 94.18, 92.18, 95.39, 91.79, 89.07, 94.72, 89.21]  # Catalyst 1 data
catalyst_2 = [89.19, 90.95, 90.46, 93.21, 97.19, 97.04, 91.07, 92.75]  # Catalyst 2 data

# Perform the independent samples t-test (assuming equal variance)
t_stat, p_value = stats.ttest_ind(catalyst_1, catalyst_2, equal_var=True)

# Print the results
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")

# Determine if the p-value is less than the significance level (alpha = 0.05)
alpha = 0.05
if p_value < alpha:
    print("Conclusion: There is a significant difference between the average yields.")
else:
    print("Conclusion: There is no significant difference between the average yields.")
```

```python
# Practical way

import math
import scipy.stats as stats

# Given values
x1 = 121  # Mean drying time for product 1
x2 = 112  # Mean drying time for product 2
sigma = 8  # Standard deviation (sigma)
n1 = n2 = 10  # Sample sizes for both groups

# Z-test calculation
z_stat = (x1 - x2) / math.sqrt((sigma ** 2 / n1) + (sigma ** 2 / n2))

# H1 Hypothesis: m1>m2

# P-value calculation for a one-tailed test
p_value = 1 - stats.norm.cdf(abs(z_stat))

# Print results
print(f"Z-Statistic: {z_stat}")
print(f"P-Value: {p_value}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Conclusion: The reduction in drying time is statistically significant.")
else:
    print("Conclusion: The reduction in drying time might be due to chance.")
```

```python
# Practical Way

import math
import scipy.stats as stats

# Given data
n1 = 500  # Sample size for exercise group
n2 = 400  # Sample size for non-exercise group
X1avg = 10.67  # Mean of exercise group
X2avg = 7.83  # Mean of non-exercise group
S1 = 3.895  # Standard deviation of exercise group
S2 = 4.224  # Standard deviation of non-exercise group

# Z-test calculation
z_stat = (X1avg - X2avg) / math.sqrt((S1**2 / n1) + (S2**2 / n2))

# Calculate the p-value for a two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Print the results
print(f"Z-Statistic: {z_stat}")
print(f"P-Value: {p_value}")

# Decision
alpha = 0.05
if p_value < alpha:
    print("Conclusion: There is a significant difference in blood pressure between the two groups.")
else:
    print("Conclusion: There is no significant difference in blood pressure between the two groups.")
```

```python
# Full code

import pandas as pd
from scipy import stats

# Load the CSV data
df = pd.read_csv('/mnt/data/cushing_data.csv')  # Update with the actual path to your CSV file

# Assuming the CSV file has columns "Group" (which indicates the disease type) 
# and "Effect" (which contains the values representing the cortisone drug effect).

# Separate the data into four groups
group_a = df[df['Group'] == 'a']['Effect']  # Adenoma
group_b = df[df['Group'] == 'b']['Effect']  # Bilateral Hyperplasia
group_c = df[df['Group'] == 'c']['Effect']  # Carcinoma
group_u = df[df['Group'] == 'u']['Effect']  # Unknown

# Perform the one-way ANOVA
f_statistic, p_value = stats.f_oneway(group_a, group_b, group_c, group_u)

# Print the results
print(f"F-Statistic: {f_statistic}")
print(f"P-Value: {p_value}")

# Decision based on the p-value
alpha = 0.05
if p_value < alpha:
    print("Conclusion: There is a significant difference between the effects of the cortisone drug on the four disease types.")
else:
    print("Conclusion: There is no significant difference between the effects of the cortisone drug on the four disease types.")

```
