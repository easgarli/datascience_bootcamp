```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import norm
```

# z- Normal Distribution

```python
# z-socre calculation

x=205
m=180
sd=20

z_score = (x-m)/sd
z_score
```

```python
# prob calculation (right - tail)

z_score = 1.25
prob = 1 - norm.cdf(z_score)
prob

# cdf: cumulative distribution function
```

```python
# prob calculation (left - tail)

z_score = 1.25
prob = norm.cdf(z_score)
prob
```

```python
# prob calculation (two - tail) (light area)

z1 = 1.25
z2 = -1.25

prob = norm.cdf(z1) - norm.cdf(z2)
prob
```

```python
# prob calculation (two - tail) (dark area)

z1 = 1.25
z2 = -1.25

prob = 1-(norm.cdf(z1) - norm.cdf(z2))
prob
```

```python
# prob calculation (dark area)

z1 = 1.25
z2 = 0

prob = norm.cdf(z1) - norm.cdf(z2)
prob
```

# t Distribution

```python
n = 15
stats.t.cdf(0, df=n-1)
```

```python
stats.t.cdf(1.75, df=n-1)
```

```python
tDist = stats.t(df=15)

x = np.linspace(-5, 5, 100)

y = tDist.pdf(x)
```

```python
plt.plot(x,y)
plt.grid()
plt.show();
```

## Calculation of p_value at t test

```python
n = 15
df = n-1
t_value = -0.7745966
p_value = stats.t.sf(abs(t_value), df)
p_value
```

```python
n = 20
df = n-1
t_value = -0.894
p_value = stats.t.sf(abs(t_value), df)*2 # 2 for two tail test
p_value
```

## Independent Sample t Test

```python
t_test=-0.35
df=14

p_value = 2*stats.t.cdf(t_test,df) # 2 for two tail
p_value
```
