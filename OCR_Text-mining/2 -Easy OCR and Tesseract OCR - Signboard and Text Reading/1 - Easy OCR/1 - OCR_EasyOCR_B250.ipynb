```python
import easyocr
import matplotlib.pyplot as plt
import cv2
```

```python
# Create a reader
reader = easyocr.Reader(['en'], gpu=False) # this needs to run only once to load the model into memory

# For Turkish
# reader = easyocr.Reader(['en', 'tr'], gpu=False)
```

```python
result = reader.readtext('London Street.png')
result
```

```python
# Perform OCR on the specified image
image_path = 'London Street.png'
result = reader.readtext(image_path)

# Load the image using OpenCV
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB for matplotlib display

# Draw bounding boxes and text on the image for each detected text area
for (bbox, text, prob) in result:
    top_left = tuple(map(int, bbox[0]))  # Top-left corner of bounding box
    bottom_right = tuple(map(int, bbox[2]))  # Bottom-right corner of bounding box
    image_rgb = cv2.rectangle(image_rgb, top_left, bottom_right, (0, 255, 0), 2)  # Draw green rectangle around detected text
    image_rgb = cv2.putText(image_rgb, text, top_left, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)  # Add text label in blue

# Display the image with annotations using matplotlib
plt.imshow(image_rgb)
plt.axis('off')  # Hide axis for cleaner display
plt.show()

#print(result)


# Print OCR results in a cleaner format
for (bbox, text, prob) in result:
    print(f"Detected Text: '{text}' with Confidence: {prob:.2f}")
```

# Original Image and Processed Image Together

```python
# Perform OCR on the specified image
image_path = 'London Street.png'
result = reader.readtext(image_path)

# Load the original image
original_image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convert for matplotlib display

# Create a copy of the image for displaying OCR results
ocr_image = image_rgb.copy()

# Draw bounding boxes and text on the OCR image
for (bbox, text, prob) in result:
    top_left = tuple(map(int, bbox[0]))  # Top-left corner of bounding box
    bottom_right = tuple(map(int, bbox[2]))  # Bottom-right corner of bounding box
    ocr_image = cv2.rectangle(ocr_image, top_left, bottom_right, (0, 255, 0), 2)  # Draw green rectangle
    ocr_image = cv2.putText(ocr_image, text, top_left, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)  # Add text label

# Display original and OCR-processed images side by side
fig, axes = plt.subplots(1, 2, figsize=(15, 10))

# Original image on the left
axes[0].imshow(image_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

# OCR processed image on the right
axes[1].imshow(ocr_image)
axes[1].set_title('Image with OCR Annotations')
axes[1].axis('off')

plt.show()

# Print OCR results in a cleaner format
for (bbox, text, prob) in result:
    print(f"Detected Text: '{text}' with Confidence: {prob:.2f}")
```

# Export the Read Text 

```python
import os

# Perform OCR on the specified image
image_path = 'London Street.png'
result = reader.readtext(image_path)

# Load the original image
original_image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convert for matplotlib display

# Create a copy of the image for displaying OCR results
ocr_image = image_rgb.copy()

# Draw bounding boxes and text on the OCR image
for (bbox, text, prob) in result:
    top_left = tuple(map(int, bbox[0]))  # Top-left corner of bounding box
    bottom_right = tuple(map(int, bbox[2]))  # Bottom-right corner of bounding box
    ocr_image = cv2.rectangle(ocr_image, top_left, bottom_right, (0, 255, 0), 2)  # Draw green rectangle
    ocr_image = cv2.putText(ocr_image, text, top_left, cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)  # Add text label

# Display original and OCR-processed images side by side
fig, axes = plt.subplots(1, 2, figsize=(15, 10))

# Original image on the left
axes[0].imshow(image_rgb)
axes[0].set_title('Original Image')
axes[0].axis('off')

# OCR processed image on the right
axes[1].imshow(ocr_image)
axes[1].set_title('Image with OCR Annotations')
axes[1].axis('off')

plt.show()

# Print OCR results with confidence levels to the console
print("OCR Results:")
for (bbox, text, prob) in result:
    print(f"Detected Text: '{text}' with Confidence: {prob:.2f}")

# Save only the detected text to a text file
output_text_filename = os.path.splitext(image_path)[0] + '.txt'
with open(output_text_filename, 'w') as f:
    for (bbox, text, prob) in result:
        f.write(text + '\n')

print(f"OCR results saved to {output_text_filename}")
```

and **paragraph=True** extracts the text as paragraphs.

# Original Text and Result Together

```python
import matplotlib.pyplot as plt
import cv2
from PIL import Image
import easyocr
import warnings

# Suppress warnings
warnings.filterwarnings("ignore")

# Initialize OCR reader
reader = easyocr.Reader(['en'], gpu=False)

# Load the original image
image_path = 'Trump.png'
image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Perform OCR on the image with paragraph formatting
results = reader.readtext(image_path, detail=1, paragraph=True)

# Display OCR results in a formatted way
print("OCR Results (Formatted):\n")
for i, (bbox, text) in enumerate(results, 1):
    print(f"Paragraph {i}:")
    print(f"{text}\n")

# Display the image and OCR results vertically
fig, ax = plt.subplots(2, 1, figsize=(10, 15))

# Display the original image
ax[0].imshow(image_rgb)
ax[0].set_title("Original Image")
ax[0].axis("off")

# Format OCR results for display
ocr_text = "\n\n".join([f"Paragraph {i}:\n{text}" for i, (_, text) in enumerate(results, 1)])
ax[1].text(0, 1, ocr_text, fontsize=12, va="top")
ax[1].set_title("OCR Results")
ax[1].axis("off")

plt.tight_layout()
plt.show()
```
