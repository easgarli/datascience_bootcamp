```python
%%writefile Streamlit_app.py
import streamlit as st
import pandas as pd
import numpy as np

# Question
st.title("Quiz App")
st.write("What is the capital of France?")

# Flag to check if the quiz is still active
quiz_active = True

# Show different screens based on correct or wrong answers
if quiz_active:
    # Buttons with four answer options
    if st.button("Berlin"):
        st.snow()  # Wrong answer
        st.error("Oops! That's not correct. Try again.")
        quiz_active = False
    elif st.button("Madrid"):
        st.snow()  # Wrong answer
        st.error("Oops! That's not correct. Try again.")
        quiz_active = False
    elif st.button("Rome"):
        st.snow()  # Wrong answer
        st.error("Oops! That's not correct. Try again.")
        quiz_active = False
    elif st.button("Paris"):
        st.balloons()  # Correct answer
        st.success("Congratulations! That's the correct answer.")
        quiz_active = False

# "Try Again" option appears when a wrong answer is selected
if not quiz_active:
    if st.button("Try Again"):
        quiz_active = True
        st.experimental_rerun()  # Reload the page to restart the quiz
```
