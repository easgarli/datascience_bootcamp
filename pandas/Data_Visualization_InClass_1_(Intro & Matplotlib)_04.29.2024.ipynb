```python
age = np.arange(25, 44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2 = [45372, 46852, 69401, 56790, 73000, 75000, 80000, 83545]
```

```python
plt.figure(figsize=(8,5))
plt.plot(age,salary, "bo", label = "SALARY")
plt.plot(age, salary_2,"r*",label = "SALARY_2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(loc = "lower right");
```

```python
plt.figure(figsize=(8,5))
plt.plot(age,salary, "bo", label = "SALARY")
plt.plot(age, salary_2,"r*",label = "SALARY_2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(bbox_to_anchor= (1,0.2));
```

```python
plt.figure(figsize=(8,5))
plt.plot(age,salary, "r", label = "SALARY")
plt.plot(age, salary_2, "y", label = "SALARY_2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(loc = "lower right");
```

```python
plt.figure(facecolor="yellow", figsize=(10,6))

ax = plt.axes()
ax.set_facecolor("xkcd:mint green")


plt.plot(age, salary, "b-", lw = 1, label = "salary 1")  # linewidth 
plt.plot(age, salary_2, "r", linestyle = ":", label = "salary 2")

plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend()
plt.show()
```

```python
plt.figure(facecolor="yellow", figsize = (10, 5))

ax = plt.axes()
ax.set_facecolor("xkcd:mint green")

plt.plot(age, salary, "b-", lw=1, label = "MAAS 1")
plt.plot(age, salary_2, "rv", ls="-", label = "MAAS 2")


plt.xlabel("AGE")
plt.ylabel("SALARY")

plt.title(label = "SALARY BY AGE")
plt.legend(bbox_to_anchor = (0.5, 0.5, 0.5, 0.5))
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(age, salary, "r")

plt.subplot(1,2,2)
plt.plot(age, salary_2, "b")
plt.tight_layout()

plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(age, salary, "r")

plt.subplot(2,1,2)
plt.plot(age, salary_2, "b")
plt.show()
```

```python
plt.figure(figsize=(20,6))
plt.subplot(2,3,1)
plt.plot(age, salary, "r")

plt.subplot(2,3,2)
plt.plot(age, salary_2, "b")


plt.subplot(2,3,3)
plt.plot(age, salary, "r") 

plt.subplot(2,3,4)
plt.plot(age, salary_2, "b")

plt.subplot(2,3,5)
plt.plot(age, salary, "r")

plt.subplot(2,3,6)
plt.plot(age, salary_2, "b")

plt.tight_layout()

plt.show()
```

## OOP Method

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(age,salary, 'r')
ax[1].plot(age,salary_2, "b")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(age,salary, 'r')
ax[1].plot(age,salary_2, "b")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(2,3)
ax[0][0].plot(age,salary, 'r')
ax[0][1].plot(age,salary_2, "b")
ax[0][2].plot(age,salary, 'r')
ax[1][0].plot(age,salary_2, "b")
ax[1][1].plot(age,salary, 'r')
ax[1][2].plot(age,salary_2, "b")
plt.tight_layout()
plt.show()
```
