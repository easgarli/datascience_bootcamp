```python
import pandas as pd

import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
tips.describe().T
```

```python
sns.scatterplot(data=tips, x='total_bill', y = 'tip');
```

```python
sns.regplot(data=tips, x='total_bill', y = 'tip');
```

```python
sns.scatterplot(data=tips, x='total_bill', y = 'tip', hue='smoker');
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=tips, x='total_bill', y = 'tip', hue='sex');
```

```python
sns.scatterplot(data=tips, x='total_bill', y = 'tip', hue='sex');
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=tips, x='total_bill', y = 'tip', style="sex", hue='day');
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = tips["total_bill"], y = tips["tip"], style=tips.sex, hue=tips.day);
```

```python
fig, ax = plt.subplots(1,2, figsize =(10,5))
sns.scatterplot(x = tips["total_bill"], y = tips["tip"], style=tips.sex, hue=tips.day, ax=ax[0])
sns.scatterplot(x = tips["total_bill"], y = tips["tip"], style=tips.smoker, hue=tips.day, ax=ax[1]);
```

```python
sns.scatterplot(x = tips["total_bill"], y = tips["tip"], style=tips.sex, hue=tips.sex, palette="viridis");
```

```python
sns.scatterplot(x = tips["total_bill"], y = tips["tip"]);
```

```python
sns.scatterplot(x = tips["total_bill"], y = tips["tip"],alpha=0.3);
```

```python
sns.scatterplot(x = tips["total_bill"], y = tips["tip"], style=tips.sex, hue=tips.sex, palette="viridis",alpha=0.5);
```

```python
sns.scatterplot(x = tips["total_bill"], 
                y = tips["tip"], 
                style=tips.sex, 
                hue=tips.sex, 
                palette="viridis",
                alpha=0.5,
                s = 50);
```

```python
sns.scatterplot(data = tips,
                x = "total_bill", 
                y = "tip", 
                style = tips.sex, 
                hue= "day", 
                palette= "cubehelix", 
                alpha= 0.7,
                s = 200);
```

```python
sns.displot(x = "total_bill", data = tips, kde = True, bins = 244);
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 6))
sns.scatterplot(x = "total_bill", y = "tip", data = tips, s = 300, hue = "sex", ax = ax[0])
sns.histplot(x = "total_bill", data = tips, kde = True, ax = ax[1])
```

```python
sns.displot(x="total_bill", data=tips, kind="hist" )
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 6))
sns.rugplot(x = "total_bill", data = tips, height=0.8, ax = ax[0])
sns.histplot(x = "total_bill", data = tips, kde = True, ax = ax[1])
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.kdeplot(data=penguins, x= "body_mass_g", hue="sex", shade=True);
```

```python
sns.kdeplot(data=tips, x= "total_bill", y="tip", shade=True);
```

```python
fig, ax = plt.subplots(2, 2, figsize = (10, 5))

sns.kdeplot(data=tips, x = "total_bill", y = "tip", color = "purple", shade = True, ax = ax[0][0])
sns.scatterplot(data=tips, x = "total_bill", y = "tip", color = "orange", ax = ax[0][1]);
sns.rugplot(x = "total_bill", data = tips, height=0.8, ax = ax[1][0])
sns.histplot(x = "total_bill", data = tips, kde = True, color = "red", ax = ax[1][1])
```

```python
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
mpg["cylinders"].nunique()
```

```python
mpg.info()
```

```python
sns.displot(x="weight", kde=True, data=mpg, hue="origin");
```

```python
sns.displot(x="weight", 
            data=mpg[mpg.cylinders.isin([4,5,8])], 
            hue="origin",
            row= "cylinders",
            col="origin",
            palette="plasma");

```

```python
sns.displot(x="weight", 
            data=mpg[mpg.cylinders.isin([4,8])], 
            hue="origin",
            row= "cylinders",
            col="origin",
            palette="viridis");

```

```python
sns.countplot(data=tips, x="day")
```

```python
sns.countplot(data=mpg, x="cylinders")
```

```python
fig, ax = plt.subplots(figsize = (10, 5))

ax_tip = sns.countplot(data = tips, x = "day", hue = "sex")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.15, p.get_height()+1))
    ax.text(p.get_x()+0.07, p.get_height()-4, "%"+str(round(p.get_height()/tips.day.count()*100, 2)));
```

```python
sns.barplot(data=tips, x ="sex", y ="tip");
```

```python
sns.barplot(data=tips, x ="sex", y ="tip",estimator=np.sum);
```

```python
sns.barplot(data=tips, x ="sex", y ="tip",estimator=np.min);
```

```python
sns.barplot(data=tips, x ="sex", y ="tip",estimator=np.max);
```

```python
sns.barplot(data=tips, x ="sex", y ="tip",estimator=np.mean);
```

```python
sns.barplot(data=tips, x ="sex", y ="tip",estimator=np.sum, ci = "sd");
```
